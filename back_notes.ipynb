# Guide for building ANN, MLP
## No of Neurons
- no of neurons in input layer = no of features in the ds
- no iof neurons is upto you or form 2^n
- no of neuron i output layer depends on task for binary classification or regression one is enouugh for multi no of classes = no of output neuron
## Activation functions for hidden layer
- In hidden layer you can use relu, leky-relu, hyperbolic-tan, selu, elu
## Activation functions for output layer
- Dependent on the task to be performed
- sigmoid(binary classification), linear(regression), softmax(multiclassification)

# Loss Functions
- Binary cross entropy, categorical cross entropy, sparse categorical cross entropy, RMSE, MSE
- Regression - RMSE, MSE, MAE
- Classification - binary cross entropy
- Multi Classification - Sparse categorical cross entropy(discrete values) or categorical cross entropy()

# Optimization Algorithms(DC)
- ADAM, SGDC, RMS PROM, NADAM

# TO build ANN
- Define the architecture(input layer, hidden layer, output layer and activation function)
- Compile the NN by telling it the loss function, optimizing performance and what metrics(RMSE, Accuracy)
- data(validation data) and no of epochs
- Model evaluation on test data

# To build the Architecture of NN

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
df = pd.read_csv(r'C:\Users\hp\Desktop\super folder\SECOND SEMESTER\UNSUPERVISED_ML\Deep Learning (2)\Deep Learning\Bank_notes.csv')
```

```python
df.head()
```

```python
X = df.drop('Class', axis=1)
Y = df.Class
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, stratify=Y, random_state=42, test_size=0.2)
```

```python
model = Sequential()
model.add(Input(shape=(4, )))
model.add(Dense(units=16, activation='elu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['Accuracy', 'precision', 'recall', 'f1_score'])
```

```python
model.fit(x_train, y_train, verbose=1, epochs=42, validation_split=1.5)
```
